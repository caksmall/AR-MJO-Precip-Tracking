## Overall Climatology (June 2000 through June 2024)
211,199 LPT Hours

```python
#plot the whole thing
fl_names = glob.glob('/home/disk/orca/csmall3/AR_testing_research/Final_AR_Results/nc_files/MJO_LPT_NC/LPT_Hours_Clim_pt*.nc')

ls_masks = []
for i in fl_names:
    ar_oi = xr.open_dataset(i)
    test1=np.array(ar_oi['MJO LPT Masks'])
    ls_masks += [test1]

plt_sum=sum(ls_masks)
```

```python
#I also need the coordinates
ar_oi = xr.open_dataset('/home/disk/orca/csmall3/AR_testing_research/Final_AR_Results/nc_files/MJO_LPT_NC/LPT_Hours_Clim_pt_4.nc')
lon_ary = ar_oi['lon']
lat_ary = ar_oi['lat']
```

```python
norm_val = 211199
plt_per = (plt_sum/norm_val)*100
```

```python
plt_per.max()
```

```python
#plot or try to 
#plot and let's see


cm = 180
colormap=cmaps.WhiteBlueGreenYellowRed
fig = plt.figure(figsize=[14, 7])
political_boundaries = NaturalEarthFeature(category='cultural', name='admin_0_boundary_lines_land', scale='50m', facecolor='none')
states = NaturalEarthFeature(category='cultural', name='admin_1_states_provinces_lakes', scale='50m', facecolor='none')

ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=cm))
ax.set_title('MJO LPT Frequency 24-Year Composite', fontsize=14)					   
ax.add_feature(cartopy.feature.LAND, edgecolor='k', facecolor='none', zorder=10)
ax.coastlines('50m', linewidth=2, zorder=5)
minlon = 70 + cm
maxlon = -130 + cm

ax.set_extent([minlon, maxlon, -30, 75], ccrs.PlateCarree())
css = ax.pcolormesh(lon_ary, lat_ary, plt_per, cmap = colormap, transform=ccrs.PlateCarree(),zorder=1, vmin=0,vmax=38.896244164169744)

cbar = plt.colorbar(css, ax=ax,orientation='horizontal',aspect=30, shrink=0.7, pad=0.06)
cbar.ax.tick_params(labelsize=16)
cbar.set_label('Percent of MJO LPT Hours', fontsize =16)


#try adding gridlines
gl = ax.gridlines(draw_labels=True, linewidth=0.5)
gl.ypadding = 5
gl.xpadding = 5
gl.xlocator = mticker.FixedLocator(np.arange(-180,180,45)[::1])
gl.ylocator = mticker.FixedLocator(np.arange(-90,90,45)[::1])
fig.text(0.50, 0.02,'This composite constitutes 211,199 hours from June 2000 through June 2024', horizontalalignment='center', wrap=True )		
fig.savefig("/home/disk/orca/csmall3/AR_testing_research/Final_AR_Results/Figures/MJO_LPT_Distribution/MJO_LPT_Distribution_Climatology.png", dpi=350, bbox_inches='tight') 
fig.show()
```

```python
#set up MJO-Connected
fl_names = glob.glob('/home/disk/orca/csmall3/AR_testing_research/Final_AR_Results/nc_files/MJO_LPT_NC/DJFM_MJO_Conn_LPT_Hours_pt_*.nc')

ls_masks = []
for i in fl_names:
    ar_oi = xr.open_dataset(i)
    test1=np.array(ar_oi['MJO LPT Masks'])
    ls_masks += [test1]

mjo_sum=sum(ls_masks)

#I also need the coordinates
ar_oi = xr.open_dataset('/home/disk/orca/csmall3/AR_testing_research/Final_AR_Results/nc_files/MJO_LPT_NC/DJFM_MJO_Conn_LPT_Hours_pt_22.nc')
lon_ary = ar_oi['lon']
lat_ary = ar_oi['lat']

norm_val = 15571
plt_mjo = (mjo_sum/norm_val)*100
```

```python
#set up the climatolgy for a difference field
#set up climatology
#plot the whole thing
fl_names = glob.glob('/home/disk/orca/csmall3/AR_testing_research/Final_AR_Results/nc_files/MJO_LPT_NC/DJFM_LPT_Hours_Clim_pt_*.nc')

ls_masks = []
for i in fl_names:
    ar_oi = xr.open_dataset(i)
    test1=np.array(ar_oi['MJO LPT Masks'])
    ls_masks += [test1]

clim_sum=sum(ls_masks)

#I also need the coordinates
ar_oi = xr.open_dataset('/home/disk/orca/csmall3/AR_testing_research/Final_AR_Results/nc_files/MJO_LPT_NC/DJFM_LPT_Hours_Clim_pt_58.nc')
lon_ary = ar_oi['lon']
lat_ary = ar_oi['lat']

norm_val = 69801
plt_clim = (clim_sum/norm_val)*100
```

```python
#do the subtraction
diff_per = plt_mjo - plt_clim
```

```python
diff_per.min()
```

```python
diff_per.max()
```

```python
#plot or try to 
#plot and let's see
from scipy.ndimage import gaussian_filter

cm = 180
# colormap=cmaps.WhiteBlueGreenYellowRed
colormap = cmaps.MPL_BrBG
fig = plt.figure(figsize=[14, 7])
political_boundaries = NaturalEarthFeature(category='cultural', name='admin_0_boundary_lines_land', scale='50m', facecolor='none')
states = NaturalEarthFeature(category='cultural', name='admin_1_states_provinces_lakes', scale='50m', facecolor='none')

ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=cm))
ax.set_title('DJFM MJO LPT Frequency Difference (MJO-Connected vs. 24-Year Composite)', fontsize=14)					   
ax.add_feature(cartopy.feature.LAND, edgecolor='k', facecolor='none', zorder=25)
ax.coastlines('50m', linewidth=2, zorder=25)
minlon = 70 + cm
maxlon = -130 + cm

ax.set_extent([minlon, maxlon, -40, 75], ccrs.PlateCarree())
css = ax.pcolormesh(lon_ary, lat_ary, diff_per, cmap = colormap, transform=ccrs.PlateCarree(),zorder=21, vmin=-20,vmax=20,alpha=.9)
# ax.contourf(lon_ary, lat_ary, fin_sig_array,levels=[0,1],hatches=[None,'..'],cmap='Greys',extend='both',alpha=0.4,zorder=20,transform=ccrs.PlateCarree())
cdssss=ax.contour(lon_ary, lat_ary, gaussian_filter(plt_clim, 2, mode='reflect'), levels=np.arange(5,int(plt_clim.max()), 10), colors='gray', linewidths=1.5, transform=ccrs.PlateCarree(),linestyle='--', zorder=21)

cbar = plt.colorbar(css, ax=ax,orientation='horizontal',aspect=30, shrink=0.7, pad=0.06, extend='max')
cbar.ax.tick_params(labelsize=16)
cbar.set_label('Percent Difference in MJO LPT Hours', fontsize =16)


#try adding gridlines
gl = ax.gridlines(draw_labels=True, linewidth=0.5,zorder=25, alpha = .7)
gl.ypadding = 5
gl.xpadding = 5
gl.xlocator = mticker.FixedLocator(np.arange(-180,180,30)[::1])
gl.ylocator = mticker.FixedLocator(np.arange(-90,90,30)[::1])

plt.clabel(cdssss, inline = True,
           fontsize=13, fmt='%1.2f', colors = 'gray')
fig.text(0.50, 0.02,'105 AR Systems and 15,571 LPT hours represented from June 2000 through June 2024', horizontalalignment='center', wrap=True )		
fig.savefig("/home/disk/orca/csmall3/AR_testing_research/Final_AR_Results/Figures/MJO_LPT_Distribution/DJFM_MJO_LPT_Freq_MJO_Conn_Anom.png", dpi=350, bbox_inches='tight') 
fig.show()
```
